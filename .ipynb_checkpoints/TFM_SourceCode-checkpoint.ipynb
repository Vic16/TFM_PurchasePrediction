## 3.2 Sesiones

```python
df_sesiones = pd.read_csv('data/Linea1-Sesiones.txt', sep=";", decimal=",", encoding="ISO-8859-1")

df_sesiones.head()
```

- **totalDays**, es el número de días que la personas entró al portal de comercio electrónico
- **totalSesiones**, es la sumatioria de sesiones que el usuario ha realizado en los días que navegó en el portal web

```python
df_sesionesTotales = df_sesiones.groupby(["IDUSUARIO"]).agg({'FECHA_SESION':'count', 
                                           'SESIONES':'sum'}).reset_index().rename(columns={'FECHA_SESION':'totalDays',
                                                                                           'SESIONES': 'totalSesiones'}
                                                                                  )
df_sesionesTotales.head(2)
```

Transformamos el campo FECHA_SESION a objeto fecha

```python
df_sesiones["FECHA_Fix"] = [datetime.strptime(i , "%d/%m/%Y %H:%M:%S") for i in df_sesiones["FECHA_SESION"]]
df_sesiones.head(2)
```

Buscamos la fecha máxima y mínima por usuario. Es decir la primera y última vez que casa usuario navegó dentro del sitio web de comercio electrónico

```python
df_sesionesFechaMin = df_sesiones[["IDUSUARIO",
                                   "FECHA_Fix"]].groupby("IDUSUARIO").agg({'FECHA_Fix':
                                                                           'min'}).reset_index().rename(columns = {"FECHA_Fix":"FECHA_Min"})


df_sesionesFechaMax = df_sesiones[["IDUSUARIO",
                                   "FECHA_Fix"]].groupby("IDUSUARIO").agg({'FECHA_Fix':
                                                                           'max'}).reset_index().rename(columns = {"FECHA_Fix":"FECHA_Max"})
```

Calculamos los deltas o diferencias entre la fecha máxima o última visista y la fecha mínima o primera visita

```python
deltas = []
for i,e in zip(list(df_sesionesFechaMax["FECHA_Max"]), list(df_sesionesFechaMin["FECHA_Min"])):
    delta = (i - e).days
    deltas.append(delta)
print(len(deltas))        
```

```python
df_sesionesTotales["DeltaDateSesiones"] = deltas
df_sesionesTotales.head()     
```

```python
df_sesionesTotales = pd.merge(df_sesionesTotales, df_sesionesFechaMin, how="left", left_on="IDUSUARIO", right_on="IDUSUARIO")     
df_sesionesTotales = pd.merge(df_sesionesTotales, df_sesionesFechaMax, how="left", left_on="IDUSUARIO", right_on="IDUSUARIO")     
df_sesionesTotales.head(2)
```

Realizamos el join con la tabla usuarios y borramos los objetos que no se utilizarán

```python
df_users = pd.merge(df_users, df_sesionesTotales, how="inner", left_on="IDUSUARIO", right_on="IDUSUARIO")
```

```python
del df_sesionesFechaMin, df_sesionesFechaMax, df_sesionesTotales, df_sesiones
```

Calculamos el número de caracteres que tiene el mail de registro de cada usuario

Realizamos varios cálculos:
- **FinalPartDominio**, se refiere a la última parte del dominio, por ejemplo: .com, .gov, .edu
- **NumberPartsDominio**, contamos el número de partes que tiene el dominio, separadas estas por cada punto dentro del string
- **LenFinalPartDominio**, contamos el número de caracteres que tiene cada parte final del dominio, se esperaría que un dominio válido o real tendría entre 3 y 2 caracteres: .com
- **NumberDotsDominio**, contamos el número de puntos (".") que tiene el dominio

```python
df_users["FinalPartDominio"] = [i.split(".")[-1] for i in df_users["TIPOEMAIL"]]
df_users["NumberPartsDominio"] = [len(i.split(".")) for i in df_users["TIPOEMAIL"]]

df_users["LenFinalPartDominio"] = [len(i) for i in df_users["FinalPartDominio"]]
df_users["NumberDotsDominio"] = [i.count(".") for i in df_users["TIPOEMAIL"]]
```

La siguiente función toma un conjunto de datos y realiza el overasampling con diferentes parámetros: iguala las clases o realiza un oversampling de la clase minoritaria de forma que representa un 33% y un 15% de la clase mayoritaria. 

Este procedimiento se realiza con tres algoritmos: 
- ROS, Random Oversampling
- SMOTE, Synthetic Minority Oversampling Technique
- ADASYN, Adaptive Synthetic

```python
def getOverSamples(data, sufix):
    """
    """
    X = data.loc[:, data.columns != 'Y']
    y = data['Y']
    
    
    ############
    ############ Random Over Sample
    ############
    
    ### Minority class to 15% on the other class 
    
    
    ros_15 = RandomOverSampler(random_state=43, sampling_strategy=.15)
    X_ros_15, y_ros_15 = ros_15.fit_resample(X, y)
    data_ros_15 = pd.DataFrame(X_ros_15)
    data_ros_15["Y"] = y_ros_15
    
    
    ### Minority class to 33% on the other class 
    
    ros_33 = RandomOverSampler(random_state=43, sampling_strategy=.33)
    X_ros_33, y_ros_33 = ros_33.fit_resample(X, y)
    data_ros_33 = pd.DataFrame(X_ros_33)
    data_ros_33["Y"] = y_ros_33
    
    ### Equal class
    ros_equal = RandomOverSampler(random_state=43, sampling_strategy=1.0)
    X_ros_equal, y_ros_equal = ros_equal.fit_resample(X, y)
    data_ros_equal = pd.DataFrame(X_ros_equal)
    data_ros_equal["Y"] = y_ros_equal
    
    
    ############
    ############ SMOTE
    ############
    
    X_smote_15, y_smote_15 = SMOTE(random_state=42, sampling_strategy=.15).fit_resample(X, y)
    data_smote_15 = pd.DataFrame(X_smote_15)
    data_smote_15["Y"] = y_smote_15
    
    X_smote_33, y_smote_33 = SMOTE(random_state=42, sampling_strategy=.33).fit_resample(X, y)
    data_smote_33 = pd.DataFrame(X_smote_33)
    data_smote_33["Y"] = y_smote_33
    
    X_smote_equal, y_smote_equal = SMOTE(random_state=42, sampling_strategy=1.0).fit_resample(X, y)
    data_smote_equal = pd.DataFrame(X_smote_equal)
    data_smote_equal["Y"] = y_smote_equal
    
      
    ############
    ############ ADASYN
    ############
    
    X_adasyn_15, y_adasyn_15 = ADASYN(random_state=42, sampling_strategy=.15).fit_resample(X, y)
    data_adasyn_15 = pd.DataFrame(X_adasyn_15)
    data_adasyn_15["Y"] = y_adasyn_15
    
    X_adasyn_33, y_adasyn_33 = ADASYN(random_state=42, sampling_strategy=.33).fit_resample(X, y)
    data_adasyn_33 = pd.DataFrame(X_adasyn_33)
    data_adasyn_33["Y"] = y_adasyn_33
    
    X_adasyn_equal, y_adasyn_equal = ADASYN(random_state=42, sampling_strategy=1.0).fit_resample(X, y)
    data_adasyn_equal = pd.DataFrame(X_adasyn_equal)
    data_adasyn_equal["Y"] = y_adasyn_equal
    
    return [
        
            ("ROS 15% {}".format(sufix),data_ros_15),
            ("ROS 33% {}".format(sufix),data_ros_33),
            ("ROS Equal Class {}".format(sufix),data_ros_equal),
        
            ("SMOTE 15% {}".format(sufix),data_smote_15),
            ("SMOTE 33% {}".format(sufix),data_smote_33),
            ("SMOTE Equal Class{}".format(sufix),data_smote_equal),
        
            ("ADASYN 15% {}".format(sufix),data_adasyn_15),
            ("ADASYN 33% {}".format(sufix),data_adasyn_33),
            ("ADASYN Equal Class {}".format(sufix),data_adasyn_equal)          
    
          ]
    
```

Aplicamos la función sobre los conjuntos de datos reducidos con PCA, UMAP y TSNE

```python
dataFrames_PCA = getOverSamples(PCA_data, "PCA")
dataFrames_UMAP = getOverSamples(UMAP_data, "UMAP")
dataFrames_TSNE = getOverSamples(TSNE_data, "TSNE")
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from sklearn import svm
from sklearn.ensemble import AdaBoostClassifier

from sklearn.model_selection import GridSearchCV
from sklearn import metrics
```

Definimos un diccionario por algoritmo de clasificación, con los parámetros a probar

```python
############### LR
logistcRegresion_params = {'C': [0.01,0.1,1,10,100,1000]}

############### DT

decisionTree_params = {"criterion":["gini","entropy"],
                       "max_depth":[5,10]    
                      }

############### RF
randomForest_params = {
    'max_depth': [5,10],
    "criterion":["gini","entropy"],
    'max_features': ['auto', 'sqrt'],
    'n_estimators': [30,100],
    'n_jobs':[-1]
    
}

############### XGB
xgb_params =    {
            'n_estimators': [30,100],
            "criterion":["gini","entropy"],
            'max_depth':[5,10],
            'learning_rate': [0.1,1.0],
            'verbosity':[0],
            'n_jobs':[-1],
            #'tree_method':["gpu_hist"]
            'gamma': [0,1],
            #'min_child_weight': [0, 0.5, 1],
            #'max_delta_step': [0],
            #'subsample': [0.7, 0.8, 0.9, 1],
            #'colsample_bytree': [0.6, 0.8, 1],
            #'colsample_bylevel': [1],
            #'reg_alpha': [0, 1e-2, 1, 1e1],
            #'reg_lambda': [0, 1e-2, 1, 1e1],
            #'base_score': [0.5]
            }



############### ADA Boost 
adaBoost_params = {
    
    'n_estimators': [30,100],
    'learning_rate': [0.1, 1.0, 1.5]
    
}
```

Definimos un objeto por algoritmo

```python
logistcRegresion = LogisticRegression()
decisionTree = DecisionTreeClassifier()
randomForest = RandomForestClassifier()
xgb = XGBClassifier()
adaBoost = AdaBoostClassifier()
```

Definimos 4 listas: los modelos, los parámetros, los nombres de los modelos y los conjuntos de datos

```python
models = [
          logistcRegresion, 
          decisionTree,
          randomForest, 
          xgb, 
          adaBoost
          ]
params = [
          logistcRegresion_params, 
          decisionTree_params, 
          randomForest_params, 
          xgb_params,
          adaBoost_params
         ]

modelname = [
             "LogisticRegresion", 
             "DecisionTree", 
             "RandomForest", 
             "XGB", 
             "AdaBoost"
             ]

data = totalDFs
```

Ejecutamos un loop en el cual por cada conjunto de datos se realiza un segundo loop donde cada modelo pasará con sus respectivos parámetros por Grid Search Cross Validation. Además los resultados son almacenados en un DataFrame al que se le agregan el nombre del modelo, el conjunto de datos y el tiempo total por bloque. 

```python
from joblib import dump, load
```

Durante la ejecución se guarda el modelo con la métrica de F1 más alta por cada grid search

```python
macroResults = []
contador = 0

for i in range(0,len(data)):

    startTime = time.time()
    for model,param, modelN in zip(models, params, modelname):
        print("Model: {}, \n Params: {},\n Dataset: {}\n".format(modelN, param, data[i][0]))
        
        model_GS = GridSearchCV(estimator=model,
                             param_grid=param, 
                             scoring = ['accuracy', 'f1','balanced_accuracy','roc_auc'],
                             n_jobs=-1,
                             cv=5,
                             verbose=False,
                             refit='f1'
                            )
        X = data[i][1].loc[:, data[i][1].columns != 'Y']
        y = data[i][1]["Y"]
        
        model_GS.fit(X,y)
        
        estimator = model_GS.best_estimator_
        dump(estimator, str("models/" + modelN + "_" + data[i][0] + "_" + ".joblib"))
        
        results = pd.DataFrame(model_GS.cv_results_)
        results["Model"] = str(model)
        results["Dataset"] = data[i][0]
        results["Grid_Time"] = round(time.time() - startTime,4)
        
        results = results[["mean_fit_time", 
                           "params", 
                           "mean_test_accuracy","std_test_accuracy",
                           "mean_test_f1","std_test_f1",
                           "mean_test_balanced_accuracy","std_test_balanced_accuracy",
                           "mean_test_roc_auc","std_test_roc_auc",
                           "Model","Dataset"
                          ]]
        
        nameFile = "results/"+str(modelN).replace("()","") +"_"+ str(data[i][0]).replace("%","").replace(" ","_")+str(i)+".csv"
        
        results.to_csv(nameFile, sep="|", index=False)
        macroResults.append(results)
        print("Tiempo de ejecución del bloque: {}".format(round(time.time() - startTime,4)))
        print("-------------------------------------------------------------------", contador)
        contador += 1
```

Se desarrolla la siguiente función para realizar las validaciones de los modelos de mayor performance con los datos originales sin balancear

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import f1_score
from sklearn.metrics import roc_auc_score
from sklearn.metrics import balanced_accuracy_score
from joblib import dump, load

def testModel(df, filelocation):
    X = df.loc[:, df.columns != 'Y']
    y = list(df["Y"])
    
    model = load(filelocation)
    yPred = model.predict(X)
    
    accuracy = accuracy_score(y, yPred)
    f1 = f1_score(y, yPred)
    roc = roc_auc_score(y, yPred)
    balanced_accuracy = balanced_accuracy_score(y, yPred)
    
    return (accuracy, f1, roc, balanced_accuracy)
```

```python
PCA_data = pd.read_csv("dataProd/dataPCA.csv", sep="|")
UMAP_data = pd.read_csv("dataProd/dataUMAP.csv", sep="|")
TSNE_data = pd.read_csv("dataProd/dataTSNE.csv", sep="|")
dataOriginal = pd.read_csv("dataProd/data_Original_.csv", sep="|")
```

```python
modelsDF = modelsDF[modelsDF["ModelType"]!= "XGB"]
```

```python
modelsDF = modelsDF.reset_index(drop=True)
modelsDF.head(4)
```

```python
metrics = []

for Model, data, location in zip(modelsDF["ModelName"], modelsDF["ModelDataset"], modelsDF["FileLocation"]):
    if data == "PCA":
        scores = testModel(PCA_data, location) 
    elif data == "UMAP":
        scores = testModel(UMAP_data, location)
    elif data == "TSNE":
        scores = testModel(TSNE_data, location)
    else:
        scores = testModel(dataOriginal, location)
    
    metrics.append(scores)
```
